**Converting Categorical Data to Numerical Data**

```python
# Getting unique values for all the categorical columns

print("Unique values for each column :\n")

print("Checking_Account :",data['Checking_Account'].unique())
print("Credit_History  :",data['Credit_History'].unique())
print("Purpose :",data['Purpose'].unique())
print("Saving_Account :",data['Saving_Account'].unique())
print("Present_Employment_Since :",data['Present_Employment_Since'].unique())
print("Personal_Status_and_Sex :",data['Personal_Status_and_Sex'].unique())
print("Other_Debtors :",data['Other_Debtors'].unique())
print("Property :",data['Property'].unique())
print("Other_Installment_Plans :",data['Other_Installment_Plans'].unique())
print("Housing :",data['Housing'].unique())
print("Job :",data['Job'].unique())
print("Telephone :",data['Telephone'].unique())
print("Foreign_Worker :",data['Foreign_Worker'].unique())
```

```python
# Converting categorical column data to numerical data

data['Checking_Account'] = data['Checking_Account'].map(
    {'A11': 0,
     'A12': 1,
     'A13': 2,
     'A14': 3})

data['Credit_History'] = data['Credit_History'].map(
    {'A30': 0,
     'A31': 1,
     'A32': 2,
     'A33': 3,
     'A34': 4})

data['Purpose'] = data['Purpose'].map(
    {'A40': 0,
     'A41': 1,
     'A42': 2,
     'A43': 3,
     'A44': 4,
     'A45': 5,
     'A46': 6,
     'A47': 7,
     'A48': 8,
     'A49': 9,
     'A410': 10})

data['Saving_Account'] = data['Saving_Account'].map(
    {'A61': 0,
     'A62': 1,
     'A63': 2,
     'A64': 3,
     'A65': 4})

data['Present_Employment_Since'] = data['Present_Employment_Since'].map(
    {'A71': 0,
     'A72': 1,
     'A73': 2,
     'A74': 3,
     'A75': 4})

data['Personal_Status_and_Sex'] = data['Personal_Status_and_Sex'].map(
    {'A91': 0,
     'A92': 1,
     'A93': 2,
     'A94': 3})

data['Other_Debtors'] = data['Other_Debtors'].map(
    {'A101': 0,
     'A102': 1,
     'A103': 2})

data['Property'] = data['Property'].map(
    {'A121': 0,
     'A122': 1,
     'A123': 2,
     'A124': 3})

data['Other_Installment_Plans'] = data['Other_Installment_Plans'].map(
    {'A141': 0,
     'A142': 1,
     'A143': 2})

data['Housing'] = data['Housing'].map(
    {'A151': 0,
     'A152': 1,
     'A153': 2})

data['Job'] = data['Job'].map(
    {'A171': 0,
     'A172': 1,
     'A173': 2,
     'A174': 3})

data['Telephone'] = data['Telephone'].map(
    {'A191': 0,
     'A192': 1})

data['Foreign_Worker'] = data['Foreign_Worker'].map(
    {'A201': 0,
     'A202': 1})

data
```

### ***Balancing Dataset (based on Risk)***

```python
# importing required libraries

from sklearn.utils import resample
```

```python
data['Risk'].value_counts()
```

```python
# Separating the data based on the 'Purpose' categories
categories = data['Risk'].unique()
category_data = {}
for category in categories:
    category_data[category] = data[data['Risk'] == category]

# Defining the number of samples to be upsampled/downsampled (max number of samples per purpose)
num_samples = max(data['Risk'].value_counts())

# Upsample or downsample each category
balanced_risk_data = []
for category, df in category_data.items():
    if len(df) < num_samples:
        # Upsample the category
        upsampled_data = resample(df, replace=True, n_samples=num_samples, random_state=42)
        balanced_risk_data.append(upsampled_data)
    else:
        # Downsample the category
        downsampled_data = resample(df, replace=False, n_samples=num_samples, random_state=42)
        balanced_risk_data.append(downsampled_data)

# Combine the balanced data for all categories
balanced_risk_data = pd.concat(balanced_risk_data)
```

```python
balanced_risk_data['Risk'].value_counts()
```

```python
# Getting data for X(fetaures) and Y(target) variables

X_bal = balanced_risk_data.drop('Risk', axis=1)
y_bal = balanced_risk_data['Risk']
```

### ***Random Forest Classifier***

**Model Implementation with Feature Selection**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier

# for feature selection
from sklearn.feature_selection import RFE

# for getting number of features to be selected
from sklearn.feature_selection import RFECV

# for cross validation strategy
from sklearn.model_selection import StratifiedKFold
```

```python
# Creating a base model
model_rf = RandomForestClassifier()
```

**Getting number of features to be selected**

```python
# Specifying the cross-validation strategy
cv = StratifiedKFold(n_splits=5)

# Initializing RFECV with the estimator (model), cross-validation, and scoring metric
rfecv = RFECV(estimator = model_rf, cv = cv, scoring = 'accuracy')

# Fitting RFECV on the data
rfecv.fit(X_bal, y_bal)

# Geting the optimal number of features
optimal_num_features = rfecv.n_features_

print("Optimal Number of Features:", optimal_num_features)
```

**Getting exact features to be selected from all features**

```python
# Selecting top k features using RFE
selector = RFE(estimator = model_rf, n_features_to_select = optimal_num_features)
X_new = selector.fit_transform(X_bal, y_bal)
```

```python
# Get the selected feature
selected_indices = selector.get_support(indices=True)
selected_features = X_bal.columns[selected_indices]

selected_features
```

**Loan Risk Prediction**

```python
# Loan Risk Prediction

X_rf_bal = balanced_risk_data[selected_features]
y_rf_bal = balanced_risk_data['Risk']

X_train, X_test, y_train, y_test = train_test_split(X_rf_bal, y_rf_bal, test_size=0.2, random_state=42)

# print("Training set shape:", X_train.shape, y_train.shape)
# print("Testing set shape:", X_test.shape, y_test.shape)

# Train a Random Forest classifier
model_rf.fit(X_train, y_train)

X_rf = data[selected_features]
y_rf = data['Risk']

X_train, X_test, y_train, y_test = train_test_split(X_rf_bal, y_rf_bal, test_size=0.2, random_state=42)

# Make predictions on the test set
y_pred = model_rf.predict(X_test)
```

```python
# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1score:", f1)
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

# Calculate silhouette scores for different numbers of clusters
silhouette_scores = []
max_clusters = 10  # Set the maximum number of clusters to consider

for n_clusters in range(2, max_clusters + 1):
    model = AgglomerativeClustering(n_clusters=n_clusters)
    labels = model.fit_predict(X)
    silhouette_scores.append(silhouette_score(X, labels))

# Find the optimal number of clusters with the highest silhouette score
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + 2  # Add 2 because range starts from 2

print("Optimal Number of Clusters:", optimal_num_clusters)
```

```python
# Perform hierarchical clustering
num_clusters = 4  # Define the number of clusters
clustering = AgglomerativeClustering(n_clusters=optimal_num_clusters)
clusters = clustering.fit_predict(X)

# Add the cluster labels to the data
balanced_purpose_data['ClusterLabel'] = clusters

# Analyze the clusters and recommend loan purposes
cluster_purposes = balanced_purpose_data.groupby('ClusterLabel')['Purpose'].apply(lambda x: x.value_counts().index[0])

# Print the recommended loan purposes by cluster
for cluster_label, purpose in cluster_purposes.items():
    print(f"Cluster {cluster_label}: Recommended Loan Purpose - {purpose}")
```
